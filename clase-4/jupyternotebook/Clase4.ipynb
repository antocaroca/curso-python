```python
import pandas as pd

archivo = "./bip.xlsx"
df = pd.read_excel(archivo,
sheetname=0,
header=10,
index_col=0,
keep_default_na=True
)
df.head()
```

```python
df.loc[df['ENTIDAD'] == 'Serviestado']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (18, 6)
entidades = df["ENTIDAD"].value_counts()
entidades.plot()
plt.title('Entidades BIP')
plt.ylabel('Cantidad de Entidades')
plt.xlabel('Entidad')
```

```python

plt.rcParams['figure.figsize'] = (18, 6)
entidades = df["ENTIDAD"].value_counts()
entidades.plot(kind='barh')
plt.title('Entidades BIP')
plt.ylabel('Cantidad de Entidades')
plt.xlabel('Entidad')
```

```python
df.groupby(['COMUNA','ENTIDAD']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Entidades BIP por Comuna')
plt.ylabel('Cantidad de Entidades')
plt.xlabel('Comuna')
```

```python
entidades = pd.DataFrame(df['ENTIDAD'].value_counts())
entidades.columns=['ENTIDAD']
entidades.plot(kind='pie', y='ENTIDAD')
```

```python

plt.rcParams['figure.figsize'] = (50, 10)
comunas = pd.DataFrame(df['COMUNA'].value_counts())
comunas.columns=['COMUNA']
comunas.plot(kind='pie', y='COMUNA')
plt.legend(loc="upper right")
```
